## “Variables, data types, and structures”

#### Variables are used to store values
#### Datatypes are type of data stored in  a Variable
     - Integers (int) - stores whole numbers (positive or negative)
     - Floats (float) - Floating point numbers with decimal
     - Strings (str) - strings are series of characters
     - Booleans (bool) - Logical values True or False
##### type()
     - type() function is used to output the datatype of a give value/variable

```python
# Type function usage
number = 10
name ="John"
percent = 70.5
is_active = True

# Printing type of a variable
print(type(name))

# Printing type of a value
print(type(100))

# check datatype condition
#if type(is_active).__name__== 'bool':
# OR
if type(is_active)== bool:
    print ("Yes, it is a logical value ")
else:
    print ("No it is not")
```

```python
# more examples
x = 3.14
print(type(x).__name__)   # float

y = "hello"
print(type(y).__name__)   # str

z = [1, 2, 3]
print(type(z).__name__)   # list
```

#### Structures

- Lists:  Dynamic arrays, mutable, defined using [] - square brackets
- Tuples: similar to lists but immutable, defines using () - parenthesis
- Dictionaries: collections, defined using {} - curly braces. key-value pairs,    
   mutable, uses keys to represent values
   e.g. person = {
			     "name": "Alice",
			     "age": 30,
			     "city": "New York"
			 }
        person["name"] = Alice
- sets : unordered collection, mutable ,like dictionaries uses {} - curly braces to define, but only has individual values, automatically eliminates duplicate values, suitable for set ops like union, difference, intersection etc. 
   e.g u_nums = {1,2,2,3,3,3,4,4,4} # output = {1,2,3,4}


```python
# Examples 
# 1. Lists
# Dynamic arrays, mutable, defined using []

numbers = [10, 20, 30, 40]

print(numbers)        # [10, 20, 30, 40]

numbers.append(50)    # add element
print(numbers)        # [10, 20, 30, 40, 50]

numbers[1] = 25       # modify element
print(numbers)        # [10, 25, 30, 40, 50]
```

```python
# 2. Tuples
# Dynamic arrays, mutable, defined using []
colors = ("red", "green", "blue")

print(colors)         # ('red', 'green', 'blue')

print(colors[0])      # red

#colors[1] = "yellow"   #❌ Error (cannot modify tuple)
```

```python
# 3. Dictionaries
#Key–value pairs, mutable, defined using {}

person = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}

print(person["name"])     # Alice

person["age"] = 31        # update value
print(person)

person["country"] = "USA" # add new key-value
print(person)
```

```python
# 4. Sets
# Unordered collection, mutable, no duplicates, defined using {}
u_nums = {1, 2, 2, 3, 3, 3, 4, 4, 4}

print(u_nums)   # {1, 2, 3, 4}

a = {1, 2, 3}
b = {3, 4, 5}

# Set operations example
print(a | b)   # Union → {1, 2, 3, 4, 5}
print(a & b)   # Intersection → {3}
print(a - b)   # Difference → {1, 2}
print(b - a)   # Difference → {4, 5}


```
