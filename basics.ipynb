#### Structures

- Lists:  Dynamic arrays, mutable, defined using [] - square brackets
- Tuples: similar to lists but immutable, defines using () - parenthesis
- Dictionaries: collections, defined using {} - curly braces. key-value pairs,    
   mutable, uses keys to represent values
   e.g. person¬†=¬†{
			     "name":¬†"Alice",
			     "age":¬†30,
			     "city":¬†"New¬†York"
			 }
        person["name"] = Alice
- sets : unordered collection, mutable ,like dictionaries uses {} - curly braces to define, but only has individual values, automatically eliminates duplicate values, suitable for set ops like union, difference, intersection etc. 
   e.g u_nums = {1,2,2,3,3,3,4,4,4} # output = {1,2,3,4}


```python
# Examples 
# 1. Lists
# Dynamic arrays, mutable, defined using []

numbers = [10, 20, 30, 40]

print(numbers)        # [10, 20, 30, 40]

numbers.append(50)    # add element
print(numbers)        # [10, 20, 30, 40, 50]

numbers[1] = 25       # modify element
print(numbers)        # [10, 25, 30, 40, 50]
```

```python
# 2. Tuples
# Dynamic arrays, mutable, defined using []
colors = ("red", "green", "blue")

print(colors)         # ('red', 'green', 'blue')

print(colors[0])      # red

#colors[1] = "yellow"   #‚ùå Error (cannot modify tuple)
```

```python
# 3. Dictionaries
#Key‚Äìvalue pairs, mutable, defined using {}

person = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}

print(person["name"])     # Alice

person["age"] = 31        # update value
print(person)

person["country"] = "USA" # add new key-value
print(person)
```

```python
# 4. Sets
# Unordered collection, mutable, no duplicates, defined using {}
u_nums = {1, 2, 2, 3, 3, 3, 4, 4, 4}

print(u_nums)   # {1, 2, 3, 4}

a = {1, 2, 3}
b = {3, 4, 5}

# Set operations example
print(a | b)   # Union ‚Üí {1, 2, 3, 4, 5}
print(a & b)   # Intersection ‚Üí {3}
print(a - b)   # Difference ‚Üí {1, 2}
print(b - a)   # Difference ‚Üí {4, 5}


```

### Control structures

##### Conditional statements
 - if, elif and else

##### Loops
 - for loop
 - while loop
 - break statement - control mechanism
 - continue statement - control mechanism  

```python
# if example

# Initialize the variable `age` with a value of 16
age = 21

# Check if the age is less than 18
if age < 18:
    # Print this message if the person is a minor
    print("You are a minor!")
# Check if the age is exactly 18
elif age == 18:
    # Print this message if the person just became an adult
    print("You just became an adult!")
# If age is greater than 18
else:
    # Print this message if the person is an adult
    print("You are an adult!")
```

```python
# for loop

# Used when the number of iterations is known.
for i in range(1, 6):
    print(i)
```

```python
# while loop
# Used when the condition is checked repeatedly.
count = 1

while count <= 5:
    print(count)
    count += 1
```

```python
# Loop Control Mechanisms
# Stops the loop completely when condition is met.

for i in range(1, 10):
    print(i)
    if i == 5:
        break
    
```

```python
# continue statement
for i in range(1, 6):
    if i == 3:
        continue # skips printing 3 
    print(i)
```

### Modules
- In Python, a module refers to a file containing executable code, be it functions, variables, or classes, that can be leveraged by other programs or scripts. 
- Modules are a way of organizing and segmenting code to improve code reusability and manageability. 
- By segregating functionalities into distinct modules, developers can maintain a clean and organized codebase, making it easier to debug, modify, and enhance.

#### Creating a module

- To create a module in Python, one simply needs to save the code they want to include in a file with a .py extension. 
- his file can contain functions, classes, and variables, as well as runnable code. 

#### Using a module
- Once a module is created, its functionalities can be accessed in another Python script or program by using the import statement. 

```python
# Module example, creating a math_ops.py file and importing it here
# importing entire module and calling functions using module name
import math_ops

result = math_ops.add(10,5)
print (result)
```

```python
# importing specific function from the module math_ops
from math_ops import subtract, add

result = add(10,5)
print (result)

result = subtract(10,5)
print(result)
```

#### subprocess 
- üëâ Enables the spawning of new processes, connecting to their input/output/error pipes, and obtaining their return codes.
- üëâ It is a better and safer replacement for os.system().
##### ‚ùó Why DevOps prefers subprocess

- More control than os.system
- Can capture logs
- Can handle errors cleanly
- Secure (no shell injection if used properly)

| Feature               | Meaning                      |
| --------------------- | ---------------------------- |
| `subprocess.run()`    | Run external command         |
| `capture_output=True` | Capture stdout & stderr      |
| `check=True`          | Raise error if command fails |
| `returncode`          | Command status               |
| `Popen`               | Advanced process control     |


```python
### Importing the module
import subprocess

# Run a simple command

subprocess.run(["ls"]) # Mac or Linux
# subprocess.run(["dir"], shell=True)
```

```python
# Capture command output
import subprocess

result = subprocess.run(
    ["echo", "Hello World"],
    capture_output=True,
    text=True
)

print(result.stdout)
```

```python
# Check command success (return code)
import subprocess

result = subprocess.run(["clear"])

print(result.returncode) #0 ‚Üí success , # non-zero ‚Üí error
```

```python
# Handle command failure
import subprocess

try:
    subprocess.run(["ls", "not_a_folder"], check=True)
except subprocess.CalledProcessError:
    print("Command failed")
```

```python
# Read both output and error

import subprocess

result = subprocess.run(
    ["ls", "not_a_folder"],
    capture_output=True,
    text=True
)

print("STDOUT:", result.stdout)
print("STDERR:", result.stderr)
```

#### Real-world DevOps example

```python
# Check if a service is running

import subprocess

result = subprocess.run(
    ["ps", "-ef"],
    capture_output=True,
    text=True
)

if "nginx" in result.stdout:
    print("Nginx is running")
else:
    print("Nginx is not running")
```

```python
# Pipe commands (advanced but useful)
import subprocess

p1 = subprocess.Popen(["echo", "Hello"], stdout=subprocess.PIPE)
p2 = subprocess.Popen(["wc", "-c"], stdin=p1.stdout, stdout=subprocess.PIPE)

output = p2.communicate()[0]
print(output.decode())
```

### socket module

- A library for network communication that allows you to build clients and servers using sockets.

##### üëâ Used for:

- client‚Äìserver communication
- APIs (low-level)
- monitoring tools
- distributed systems

##### üß† Simple idea 

- Server ‚Üí listens on an IP + port
- Client ‚Üí connects to that IP + port
- They send and receive data

```python
# How to run
# run server.py and client.py on two separate terminals

### server.py
import socket

# create a socket
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# bind to IP and port
server_socket.bind(("127.0.0.1", 5000))

# listen for connections
server_socket.listen(1)
print("Server is listening...")

# accept client connection
conn, addr = server_socket.accept()
print("Connected by", addr)

# send data to client
conn.sendall(b"Hello from server")

# close connection
conn.close()
server_socket.close()

###########################
# client.py

import socket

# create a socket
client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# connect to server
client_socket.connect(("127.0.0.1", 5000))

# receive data
data = client_socket.recv(1024)
print("Received:", data.decode())

# close connection
client_socket.close()
```

#### json Module

- üìå Purpose
The json module helps in encoding (Python ‚Üí JSON) and decoding (JSON ‚Üí Python) data.
This is very common in microservices, REST APIs, DevOps automation, and cloud tools.

##### Summary

|Function |	Purpose |
|--------- | ------- |
|json.dumps() |	Python ‚Üí JSON string |
| json.loads() |	JSON string ‚Üí Python |
| json.dump() |	Python ‚Üí JSON file |
| json.load() |	JSON file ‚Üí Python |

#### üìä YAML vs JSON Comparison

| Feature        | YAML                      | JSON                         |
| -------------- | ------------------------- | ---------------------------- |
| Readability    | ‚≠ê Very readable           | ‚ùå Less readable              |
| Syntax         | Indentation-based         | Braces `{}` & brackets `[]`  |
| Comments       | ‚úÖ Supported               | ‚ùå Not supported              |
| Used in DevOps | Kubernetes, Ansible, Helm | APIs, Terraform output, REST |
| Parsing speed  | Slower                    | Faster                       |
| Strictness     | Flexible                  | Very strict                  |


```python
# Importing the json module
import json

# Encoding (Python ‚Üí JSON)
# Converting Python objects into JSON format.
# Example: Dictionary to JSON string

data = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}

json_data = json.dumps(data) # dumps() ‚Üí dump string (returns JSON as string)

print(json_data)
```

```python
#Decoding (JSON ‚Üí Python)
# Converting JSON data back into Python objects.
# Example: JSON string to Dictionary

import json

json_string = '{"name": "Bob", "age": 25, "city": "London"}'

python_data = json.loads(json_string) #loads() ‚Üí load string (returns Python dict)

print(python_data)
print(python_data["name"])
```

```python
# orking with JSON files
# Writing JSON to a file

import json

config = {
    "server": "prod",
    "port": 8080,
    "debug": False
}

with open("config.json", "w") as f:
    json.dump(config, f)
```

```python
# reading JSON from a file
import json

with open("config.json", "r") as f:
    config_data = json.load(f)

print(config_data["server"])
```

```python
# Real-world (DevOps) Example
# Reading API response or config file

import json

api_response = '{"status": "success", "code": 200}'

response = json.loads(api_response)

if response["status"] == "success":
    print("API call successful")
```
